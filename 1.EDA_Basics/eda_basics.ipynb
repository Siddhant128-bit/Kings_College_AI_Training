<h4> 3.1. Learn to use apply </h4>

sometimes you might want to make some changes to the data in dataframe. Using apply saves alot of time for such operations 

<b>a. Convert all no to 0 and yes to 1 for easy processing in labels</b>

```python
def convert_to_number(x):
    if x=='no':
        return 0 
    elif x=='yes':
        return 1

print('Before Change: ')
display(labels)
print('**'*25) 

labels['play']=labels['play'].apply(convert_to_number)
print('After Change: ')
display(labels)
print('**'*25)
# YOUR CODE HERE
# Convert all sunny to 0, overcast to 1 and rainy to 2 for features table
```
